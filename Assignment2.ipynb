```python
data = pd.read_csv('raw_data.csv')
data.head(10)
```

```python
# Task-2: Removing units from the attributes
data['Mileage'] = data['Mileage'].str.extract('(\d+\.\d+)')
data['Mileage'] = data['Mileage'].astype(float)

data['Engine'] = data['Engine'].str.extract('(\d+)')
data['Engine'] = data['Engine'].astype(float)

data['Power'] = data['Power'].str.extract('(\d+\.\d+)')
data['Power'] = data['Power'].astype(float)

data['New_Price'] = data['New_Price'].str.extract('(\d+\.\d+)')
data['New_Price'] = data['New_Price'].astype(float)

data.head(10)
```

```python
# Dropping the first column which is just a row number
data = data.iloc[:, 1:]
```

```python
#Task-1:

# Identifing missing values in all columns
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Drop the New_Price column
data.drop(columns=['New_Price'], inplace=True)

# Handle missing values: For now, we'll fill missing values with the mean of the respective columns
data['Mileage'].fillna(data['Mileage'].mean(), inplace=True)
data['Engine'].fillna(data['Engine'].mean(), inplace=True)
data['Power'].fillna(data['Power'].mean(), inplace=True)
data['Seats'].fillna(data['Seats'].mean(), inplace=True)
```

```python
# Identifing missing values in all columns
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Task-3:

# Convert categorical variables to numerical one-hot encoded values
encoded_data = pd.get_dummies(data, columns=['Fuel_Type', 'Transmission'])
encoded_data.head()
```

```python
# Task-4:

import datetime as dt

# Calculating the current year
current_year = dt.datetime.now().year

# Create the new feature 'Car_Age' by subtracting 'Year' from the current year
data['Car_Age'] = current_year - data['Year']

data.head(10)
```

```python
# Task-5

# Select
selected_data = data[['Name', 'Location', 'Year', 'Mileage', 'Price']]

# Filter
filtered_data = data[data['Price'] > 25]

# Rename
renamed_data = data.rename(columns={'Year': 'Manufacture_Year', 'Price': 'Selling_Price'})

# Mutate
data['Price_Category'] = pd.cut(data['Price'], bins=[0, 5, 10, 15, 20, 25, 30], labels=['<5', '5-10', '10-15', '15-20', '20-25', '>25'])

# Arrange
arranged_data = data.sort_values(by='Year', ascending=False)

# Summarize
summary_data = data.groupby('Location')['Price'].mean()

# Displaying results
print("Selected Data:")
print(selected_data.head())

print("\nFiltered Data:")
print(filtered_data.head())

print("\nRenamed Data:")
print(renamed_data.head())

print("\nMutated Data with Price Category:")
print(data.head())

print("\nArranged Data:")
print(arranged_data.head())

print("\nSummary Data (Average Price by Location):")
print(summary_data)
```
